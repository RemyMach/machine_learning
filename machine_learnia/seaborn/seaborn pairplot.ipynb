# Exemple titanic

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# on supprime les colonnes qui ne nous intéresse pas ainsi que les lignes qui n'ont pas une valeur dans 
# chaque colonne que l'on garde
```

```python
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
# gaphique spécial car beaucoup de catégories
```

## Catplot

```python
#on va effectué un catplot en fonction de deux variable
sns.catplot(x='pclass', y='age', data=titanic)
```

```python
# maintenant on va ajoute le sex en plus de ça
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

# Boxplot

outliers = anomalie(souvent des datas qu'on sort du dataset car elle ne représente pas nos données et pourraient fausser nos résultats)


```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

# les distributions

```python
# visualiser la distribtuion de la collone fare donc la variable tarif
sns.distplot(titanic['fare'])
# on peut observer que la majorité des gens on prit les tarif les moins cher
```

```python
# distribution selon deux variables à la foix
sns.jointplot('age', 'fare', data=titanic)
```

```python
# on peut changer le style de celui-ci avec le paramètre kind
sns.jointplot('age', 'fare', data=titanic, kind='kde')
```

# Visualisation de matrice

```python
# matrice de corrélation
sns.heatmap(titanic.corr())
```
