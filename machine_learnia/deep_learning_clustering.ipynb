# Programmer un K-Mean Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
# Générer des données:
# 100 exemples à 2 features en simulant 3 clusters
X, y = make_blobs(n_samples=100, centers = 3, cluster_std=0.5, random_state=0) #nb_features = 2 par défaut
plt.scatter(X[:,0], X[:, 1])
```

```python
# Entrainer le modele de K-mean Clustering
# 2 clusters
model = KMeans(n_clusters=2)
model.fit(X)

#Visualiser les Clusters
predictions = model.predict(X)
plt.scatter(X[:,0], X[:,1], c=predictions)
```

```python
# Entrainer le modele de K-mean Clustering
# 3 clusters
model = KMeans(n_clusters=3)
model.fit(X)

#Visualiser les Clusters
predictions = model.predict(X)
plt.scatter(X[:,0], X[:,1], c=predictions)
```

```python
# Entrainer le modele de K-mean Clustering
# 4 clusters
model = KMeans(n_clusters=4)
model.fit(X)

#Visualiser les Clusters
predictions = model.predict(X)
plt.scatter(X[:,0], X[:,1], c=predictions)
```
