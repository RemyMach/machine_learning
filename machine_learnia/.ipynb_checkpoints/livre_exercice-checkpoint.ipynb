```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
```

```python
np.random.seed(0)
# x est juste un vecteur et y représentent les target data
x, y = make_regression(n_samples= 100, n_features=1, noise=10)
plt.scatter(x, y)
```

## entraîner le modèle sur 100 itérations avec un learning rate de 0.0001

```python
model = SGDRegressor(max_iter=100, eta0=0.0001)
#maintenant on entraîne ce modèle
model.fit(x, y)
```

## on peut maintenant utiliser la fonction score qui calcule le coef de détermination entre le modèle et les valeurs y de notre Dataset

```python
print('Coeff R2 =', model.score(x, y))
plt.scatter(x, y)
# on fait de nouvelles prédictions avec la fonction predict et on trace des résultats avec la fonction plt.plot
plt.plot(x, model.predict(x), c='red', lw =3)
```

## le modèle n'est pas bon du tout

```python
model = SGDRegressor(max_iter=1000, eta0=0.001)
#maintenant on entraîne ce modèle
model.fit(x, y)
```

```python
print('Coeff R2 =', model.score(x, y))
plt.scatter(x, y)
# on fait de nouvelles prédictions avec la fonction predict et on trace des résultats avec la fonction plt.plot
plt.plot(x, model.predict(x), c='red', lw =3)
```

## régression linéaire plynomiale

```python
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(0)

# création du Dataset
x, y = make_regression(n_samples=100, n_features=1, noise=10)
y = y **2 # y ne varie plus linéairement selon x

# On ajoute des variables plynômiales dans notre dataset
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x = poly_features.fit_transform(x)

plt.scatter(x[:,0],y)
x.shape #la dimension de x: 100 lignes et 2 colonnes
```

```python
# On entraine le modele comme avant ! rien ne change !
model = SGDRegressor(max_iter=1000, eta0=0.001)
model.fit(x, y)
print('Coeff R2 =', model.score(x, y))

plt.scatter(x[:,0], y, marker='o')
plt.scatter(x[:,0], model.predict(x), c='red', marker='+')
```
