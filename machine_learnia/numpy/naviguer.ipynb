# navigation dans un tableau

# L'axe 0 correspond à l'axe des ordonnées donc les lignes
# L'axe 1 correspond à l'axe des abscisse donc les colonnes

```python
import numpy as np

print(A)
```

```python
print(A[0][1])
print(A[0, 1])
```

```python
# imprimer toute la première colonne du tableau A
print(A[:, 0])
```

```python
# imprimer toute la première ligne du tableau A
print(A[0,:])
#on peut aussi le faire comme ceci
print(A[0])
```

```python
# sélectionner les deux premières lignes et les deux premières colonne de notre tableau
# le deux est le point d'arrêt donc pas prit en compte quand on fait du slicing
B = A[0:2, 0:2]
print(B)
```

```python
B[0:2, 0:2] = 10
print(B)
```

```python
print(A)
print(A[:A.shape[0], 1:A.shape[1]])
# le moins permet de partir de la fin du tableau
print(A[:, -2:])
```

```python
B = np.zeros((4, 4))
print(B)
```

```python
# remplacer ce qui est en vert par des 1
C = B[1:-1, 1:-1]
C[:]= 1
print(C)
```

```python
B[1:-1, 1:-1] = C
print(B)
```

```python
# autre méthode
B = np.zeros((4, 4))
print(B)
B[1:-1, 1:-1] = 1
print(B)
```

## rajouter le pas au slicing

```python
A[::2,::2] = 1
print(A)
```

```python
# remplir un tableau avec des valeurs aléatoire entre 0 et 10 et de dimension 5 par 5
A = np.random.randint(0, 10, [5, 5])
print(A)
```

# Boolean Indexing

```python
# tester toutes les valeurs de A
B = A < 5
print(B)
```

```python
# changer tous les éléments de A qui vont être inférieur à 5
A[A < 5] = 11
print(A)
```

```python
# tester plusieurs conditions
C = np.random.randint(0, 10, [5, 5])
print(C)
```

```python
C[(C < 5) & (C > 2)] = 11
print(C)
```

```python
# retourner un tableau avec tous les nbre inférieur à n
# ça détruit sa shape comme on peut le voir
print(C[C<11])
print(C[C<11].shape)
```

```python
# il est aussi possible d'utiliser le mask d'un autre tableau pour son tableau:
E = np.random.randn(5, 5)
print(E)
```

```python
print(E[C<11])
print(E[C<11].shape)
```

# Exercice

```python
from scipy import misc
import matplotlib.pyplot as plt
# on cahrge la photo
face = misc.face()
# on charge la photo dans matplotlib pour pouvoir l'afficher
plt.imshow(face)
plt.show()
# face est un tableau numpy:
print(type(face))
```

```python
# on a un tableau à 3 dimensions comme on peut le vérifier
# on peut le deviner facilement comme on a les trois couleurs
print(face.shape)
```

```python
# pour simplifier l'exercice on va préférer travailler sur un tableau à deux dimensions
# on cahrge la photo en noir et blanc
faceGray = misc.face(gray=True)
# on charge la photo dans matplotlib pour pouvoir l'afficher
plt.imshow(faceGray, cmap=plt.cm.gray)
plt.show()
# face est un tableau numpy:
print(type(faceGray))
# on ragarde sa shape et il a bien seulement deux dimensions
print(faceGray.shape)
```

```python
# zoomer de 1 quart vers le millieu de la photo
# puis augmenter la luminosité des pixels qui sont déjà au dessus d'une certaine luminosité
# puis réduire la luminosité des pixels qui sont déjà au dessous d'une certaine luminosité
# 255 blanc, 0 noir
shapePicture0 = faceGray.shape[0]
shapePicture1 = faceGray.shape[1]
print(shapePicture0)
print(shapePicture1)
```

```python
shapePictureQuart0 = int(shapePicture0/4)
shapePictureQuart1 = int(shapePicture1/4)
faceGrayZoom = faceGray[shapePictureQuart0:(shapePictureQuart0*2),shapePictureQuart1:(shapePictureQuart1*2)]
plt.imshow(faceGrayZoom, cmap=plt.cm.gray)
plt.show()
```

```python
# maintenant on éclaircie
faceGrayZoom[faceGrayZoom > 200] = 255
plt.imshow(faceGrayZoom, cmap=plt.cm.gray)
plt.show()
```

```python
faceGrayZoom[faceGrayZoom < 50] = 0
plt.imshow(faceGrayZoom, cmap=plt.cm.gray)
plt.show()
```

# correction

```python
h = faceGray.shape[0]
w = faceGray.shape[1]
# // pour avoir forcément un entier
zoom_face = faceGray[h//4 : -h//4, w//4: -w//4]
print(zoom_face)
```

```python
# on a les pixels qui ont déja été modifié
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
# réduire la taille de notre image en gardant un pixel sur deux
# on peut voir avec les axes que ça change
faceGrayReduce = faceGray[::2, ::2]
plt.imshow(faceGrayReduce, cmap=plt.cm.gray)
plt.show()
```
