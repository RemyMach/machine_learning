# Manipulation

```python
import numpy as np

A = np.zeros((3, 2))
B = np.ones((3, 2))

print(A)
print(B)
```

```python
# coller les tableaux ligne par ligne
#on crée un tuple, on peut le remarquer avec les parenthèses
C = np.hstack((A, B))
E = np.concatenate((A, B), axis=1)
print(C)
print(E)
print(C.shape)
```

```python
# coller les tableaux l'un à la suite de l'autre
D = np.vstack((A, B))
F = np.concatenate((A, B), axis=0)
print(D)
print(F)
print(D.shape)
```

```python
# redimensionnement d'un tableau avec un nombre d'éléments égaux
np.random.seed(3120)
array_exemple = np.random.randn(2,6)
print(array_exemple)
```

```python
# remaniement d'un tableau
array_exemple = array_exemple.reshape((3,4))
print(array_exemple)
```

```python
np_list = np.array([1,2,3])
print(np_list.shape)
```

```python
# cette affichage d'un 3 sans rien après peut poser problème donc:
np_list = np_list.reshape((np_list.shape[0], 1))
np_list.shape
```

```python
# squeeze fait disparaitre toutes les dimesnions dans lesquels on a un 1
np_list = np_list.squeeze()
print(np_list.shape)
```

```python
# applatir les dimensions du tableau pour avoir un tableau à une dimension
np_array_exemple = np.random.randn(3,4)
print(np_array_exemple)
np_array_exemple = np_array_exemple.ravel()
print(np_array_exemple)
print(np_array_exemple.shape)
```

```python
def initialisation(m, n):
    # m : nombre de lignes
    # n : nombre de colonnes
    # retourne une matrice alatoire(m, n+1)
    # avec une colonne biais (remplie de "1") tout à l'heure
    array = np.random.randn(m ,(n))
    ones = np.ones((m,1))
    result = np.concatenate((array, ones), axis=1)
    
    return result
    
pomme = initialisation(3,4)
print(pomme)
```

## correction

```python
def initialisation(m, n):
    # m : nombre de lignes
    # n : nombre de colonnes
    # retourne une matrice alatoire(m, n+1)
    # avec une colonne biais (remplie de "1") tout à l'heure
    array = np.random.randn(m ,(n))
    result = np.concatenate((array, np.ones((array.shape[0], 1))), axis=1)
    
    return result
    
pomme = initialisation(3,4)
print(pomme)
```
