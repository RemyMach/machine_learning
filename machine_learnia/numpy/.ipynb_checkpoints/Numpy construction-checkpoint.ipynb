#### avec excel on travail avec des tableaux à deux dimensions, la ligne représente la première dimension et la colonne à l'intérieur de de ceille ci la deuxieme [ligne][colonne]

## array.shape nous donne la forme de notre tableau numpy , ligne, colonne et même profondeur

```python
import numpy as np

#tableau à une dimension
#on obtient un objet avec une shape de type tuple
A = np.array([1,2,3])
print(A)
print(type(A.shape))
```

```python
#nbre de dimension
print(A.ndim)
```

```python
#forme
#ici sur notre unique dimension on a 3 éléments
A.shape
```

```python
# initialiser notre tableau avec pleins de zéros
B = np.zeros((3,2))
print(B)
print(B.ndim)
print(B.shape)
```

```python
# initialiser notre tableau avec pleins de zéros
C = np.ones((3,2))
print(C)
print(C.ndim)
print(C.shape)
# taille du tableau donc nombre d'élément dans ce tableau 3*2=6
print(C.size)
```

```python
#remplir un tableau en lui donnant la forme que l'on veut avec le chiffre que l'on veut
D = np.full((2,3), 9)
print(D)
print(D.ndim)
print(D.shape)
```

```python
# initialiser un tableau numpy avec des nombres aléatoire celon la distribution normal centré en zéro
np.random.randn(3, 4)
```

```python
#on peut aussi fixer notre aléatoire pour que ce soit toujours le même en utilisant la fonction seed avant
np.random.seed(3120)
np.random.randn(3, 4)
```

```python
#construire une matrice identité avec eye()
np.eye(4)
```

```python
# construire un tableau à une dimension avec en paramètre le premier nombre, 
# le dernier nombre non compris et l'espace entre chaque éléments

np.arange(0, 10, 2)
```

```python
# construire un tableau à une dimension avec en paramètre le premier nombre, 
# le dernier nombre et le nbr d'éléments

np.linspace(0, 10, 8)
```

```python
# on peut préciser le type et la taille des éléments dans ce tableau
np.linspace(0, 10, 8, dtype=np.float16)
```
