### on peut additionner des tableaux mais il faut qu'il soit de même dimensions

# le broadcasting revient à étendre les dimensions d'un tableau

```python
import numpy as np
A = np.random.randint(0, 10, [3,3])
B = np.random.randint(0, 10, [3,3])
C = np.random.randint(0, 10, [3,2])
print(A)
print(B)
```

```python
# possible car deux matrices de même dimensions
A + B
```

```python
# impossible pas les mêmes dimesnions
A + C
```

```python
# possible car c'est juste un scalaire donc on ajoute 2 à chaque élément de A
A + 2
```

```python
A = np.random.randint(0, 10, [3,3])
B = np.random.randint(0, 10, [3,1])
print(A)
print(B)
```

```python
#ça fonctionne comme on peut le voir sur la capture
A+B
```

```python
A = np.random.randint(0, 10, [4,1])
B = np.random.randint(0, 10, [1,3])
print(A)
print(B)
```

```python
print(A+B)
```
