# Interpolate

ici on peut voir qu'il nous manque des valeurs mais nos donnes quand mme tre utilis, il nous suffit d'interpoler
les valeurs q'ils manque. Quand on regarde les colonnes, elles semble tre linaire

```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 30)
y = x**2
plt.scatter(x,y)
plt.show()
```

```python
# cration de la fonction interpolation f
f = interp1d(x, y, kind='cubic')

# rsultats de la fonction interpolation f sur de nouvelles donnes
new_x = np.linspace(0, 10, 50)
result = f(new_x)

# visualisation avec matplotlib
plt.scatter(x, y)
plt.plot(new_x, result, c='r')
```
