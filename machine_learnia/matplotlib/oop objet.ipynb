# OOP version non recommandé

```python
import matplotlib.pyplot as plt
import numpy as np
# créationde la figure et de l'axe qui snt des objets
# plt.subplots différent de plt.subplot
fig, ax = plt.subplots()
x = np.linspace(0,10,20)
y =x**2
ax.plot(x, y)
plt.show()
```

```python
# créer des subplots avec oop
fig, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(x, y)
ax[1].plot(x, y**2)
plt.show()
```
