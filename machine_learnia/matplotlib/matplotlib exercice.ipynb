# Exercice

```python
import numpy as np
import matplotlib.pyplot as plt
dataset = {f'experience{i}': np.random.randn(100) for i in range(4)}
print(dataset)
```

```python
def graphique(dataset):
    compteur = 1
    for i in dataset:
        y = []
        x = []
        for key, value in enumerate(dataset[i]):
            y.append(value)
            x.append(key)
        plt.figure(figsize=(12,8))
        plt.subplot(4, 1, compteur)
        plt.plot(x, y)
        plt.show()
        compteur += 1

graphique(dataset)
```

# Corrigé

```python
#zip premet l'itérationde deux structures de données en parrallèle
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k,i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
    
graphique(dataset)
```
