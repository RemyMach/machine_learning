# 1) graphique de classification scatter

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
# signifie qu'on éxecute et qu'on visualise les graphes dans jupyter
%matplotlib inline
# exécuter dans une fenêtre à part
# %matplotlib
```

# pour décortiquer le dataset, chaque colonne va représenter une variable exemple longueur sépal et chaque ligne va représenter ces variable d'une fleur, donc avec une ligne on a toutes les données d'une fleur qui sont ici au nombre de 4, y représente bien évidemment le résultat si ce sont ou non des fleurs de la classe 0, 1, 2 avec ces mêmes valeurs. Donc la fleur à la ligne 0 et de la classe 0

```python
iris = load_iris()
x = iris.data
y = iris.target
names = list(iris.target_names)
print(x.shape)
print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y\'a {np.unique(y).size} classes')
print(y)
```

```python
# on prend deux des 4 variables (colonnes) dont on a besoin, et les 150 lignes et on leur définit une couleur enfonction de la valeur
# de y qui correspond
plt.scatter(x[:, 0], x[:, 1], c=y)
#nos deux variables sont la longueur du sépal et la largeur de celui-ci
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.show()
# on observe très bien que en fonction de la valeur de ces deux paramètre 
#  nos valeurs sont plutôt d'une classe ou d'une autre
```

```python
# paramètre alpha permet de controller la transparence des points
# paramètre s qui permet de controller la taille de nos points
# pour s on utilise la 3 eme colonne dans notre tableau pour faire grossir nos fleurs en fonction de celle-ci
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:,2]*100)
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.show()
```

# 4) Contour plots (très utile en IA)

```python
# lambda en python générateur de fonction anonyme
f = lambda x,y: np.sin(x) + np.cos(x+y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
# obligatoire pour créer la forme de tableau que l'on veut pour le graphique
X, Y = np.meshgrid(X, Y)
# on fait du broadcasting sur les éléments de X et Y avec la fonction anonyme
Z = f(X, Y)

plt.contour(X ,Y, Z )
```

```python
# donc on peut par exmple changer le nombre de niveau et la couleur
# ligne continu positif et pointillé négatif
plt.contour(X ,Y, Z, 20, colors='black' )
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
# la color bar nous indique les valeurs de nos gradients, ce qui eput nous aider à trouver le minimum globale
# dans un problème de stochasique gradient descent
```

# exercice affiché les graphes des variables de l'iris avec l'aide de la fonction

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k,i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```
